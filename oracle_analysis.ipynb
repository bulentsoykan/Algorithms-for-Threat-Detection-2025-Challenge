```python
import matplotlib.pyplot as plt
import numpy as np
import os
import tqdm

import atd2025
```

Check accuracy of various oracle models

```python
folder = "oracle_out"
accs = []
ks = []

for filename in tqdm.tqdm(os.listdir(folder)):
    ks.append(int(filename.split("_")[1].split(".")[0]))
    filename = os.path.join(folder, filename)
    accs.append(atd2025.accuracy.evaluate_predictions(filename, "https://www.maserv.work/ATD/model2/ucf_atd_model/datasets/dataset1_truth.csv"))
```

```python
accs = np.array(accs)
ks = np.array(ks)

sort = np.argsort(ks)
ks = ks[sort]
accs = accs[sort]
```

```python
plt.plot(ks[:-1], accs[:-1])
plt.title("Oracle Accuracy for Various K")
plt.xlabel("K")
plt.ylabel("Posit Accuracy")
```
